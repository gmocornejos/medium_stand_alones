```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

from sgd_L1 import NeuralNetwork

rng = np.random.default_rng()

with open("../data/train_data.pkl", "rb") as train_file:
    train_data = pickle.load(train_file)

with open("../data/test_data.pkl", "rb") as test_file:
    test_data = pickle.load(test_file)
```

```python
# train without L1 regularization

nn = NeuralNetwork(
    [28*28, 1024, 512, 128, 10], # layers size
    1e-3,                        # learning rate
    0,                           # L1 lambda
    64,                          # mini batch size
    100                           # training epochs
)     

acc_no_reg, loss_no_reg = nn.train(train_data, test_data)
```

```python
# train with small L1 regularization

nn = NeuralNetwork(
    [28*28, 1024, 512, 128, 10], # layers size
    1e-3,                        # learning rate
    1e-1,                        # L1 lambda
    64,                          # mini batch size
    25                           # training epochs
)

acc_L1, loss_L1 = nn.train(train_data, test_data)
```

```python
plt.plot(loss_no_reg[1:], label="No reg")
plt.plot(loss_L1[1:], label="L1 reg")
plt.legend()
```

```python
plt.plot(acc_no_reg[1:], label="No reg")
plt.plot(acc_L1[1:], label="L1 reg")
plt.legend()
```
